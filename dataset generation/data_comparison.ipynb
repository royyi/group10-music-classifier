```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA

import sys

sys.path.append('../')

from eda import util
```

```python
tracks = util.load('../data/fma_metadata/tracks.csv')
features = util.load('../data/fma_metadata/features.csv')
small_rock_features = util.load('small_rock_features.csv')
small_elec_features = util.load('small_elec_features.csv')
```

```python
small_rock_features.dropna(axis=0, how='any', inplace=True)
small_elec_features.dropna(axis=0, how='any', inplace=True)
```

```python
feature_names = features.columns.levels[0]
```

```python
small = tracks['set', 'subset'] <= 'small'
elec = tracks['track', 'genre_top'] == 'Electronic'
rock = tracks['track', 'genre_top'] == 'Rock'
```

```python
import warnings

warnings.filterwarnings("ignore")


for feat in feature_names:
    fig = plt.figure(figsize=(12,8))
    fig.suptitle(feat)

    X_elec_1 = small_elec_features.loc[small & elec, feat]
    X_elec_1 = PCA(n_components=2).fit_transform(X=X_elec_1)
    X_elec_2 = features.loc[small & (elec), feat]
    X_elec_2 = PCA(n_components=2).fit_transform(X=X_elec_2)

    plt.scatter(X_elec_1[:,0], X_elec_1[:,1], color='red', marker=',')
    plt.scatter(X_elec_2[:,0], X_elec_2[:,1], color='blue', marker=',')

    plt.show()
```

```python



for feat in feature_names:
    fig = plt.figure(figsize=(12,8))
    fig.suptitle(feat)


    X_rock_1 = small_rock_features.loc[small & rock, feat]
    X_rock_1 = PCA(n_components=2).fit_transform(X=X_rock_1)
    X_rock_2 = features.loc[small & rock, feat]
    X_rock_2 = PCA(n_components=2).fit_transform(X=X_rock_2)

    plt.scatter(X_rock_1[:,0], X_rock_1[:,1], color='red', marker=',')
    plt.scatter(X_rock_2[:,0], X_rock_2[:,1], color='blue', marker=',')

    plt.show()
```
